```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
data = load_digits()
```

```python
X = data.data
y = data.target
print("Image Data shape",X.shape)
print("Label Data shape",y.shape)
```

```python
plt.figure(figsize=(20,4))
for index,(image,label) in enumerate(zip(data.data[0:5],data.target[0:5])):
    # plt.plot(1,5,index+1)
    plt.plot(1,6,index+1)
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label,fontsize=20)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(X_train.shape)
```

```python
print(y_train.shape)
```

```python
print(X_test.shape)
```

```python
print(y_test.shape)
```

```python
logr = LogisticRegression()
logr.fit(X_train,y_train)
```

```python
print(logr.predict(X_test[0].reshape(1,-1)))
```

```python
print(logr.predict(X_test[0:20]))
```

```python
predictions = logr.predict(X_test)
predictions
```

```python
score = logr.score(X_test,y_test)
print(score)
```

```python
#Evaluating with confusion matrix
cm = confusion_matrix(y_test,predictions)
print(cm)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot=True,fmt=".3f",linewidths=.5,square=True,cmap='Blues_r')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = "Accuracy score {}".format(score*100) + "%"
plt.title(all_sample_title,size=15)
```
